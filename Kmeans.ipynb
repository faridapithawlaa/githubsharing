```python
#inporting the libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

#loading the data
data = pd.read_csv("Mall_Customers.csv")

#printing the fisrt five rows of data
data.head()

#reading the data for count, mean etc.
data.describe()

#calculating number of male and females
male= data[data['Gender']== 'Male']
female= data[data['Gender']== 'Female']

#plotting pie chart for the percentage of males ad females
plt.pie([len(male), len(female)],labels=['Male', 'Female'], colors=["r", "g"],autopct='%0.2f%%', shadow=True, startangle=140)
plt.show()

#plotting bar graph to show varition in the number of male and females
plt.figure(figsize=(10,7))
sns.countplot(data=data,x=data['Gender'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

#historgram shawing the count of ages
sns.distplot(a=data['Age'] ,kde=False)
plt.xlabel('Ages')
plt.ylabel('Frequency')
plt.show()



#encoding the string data type of attributes
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
for i in range(4):
    data.iloc[:,i]=le.fit_transform(data.iloc[:,i])
    
#importing the kmeans library
from sklearn.cluster import KMeans

#calculating and printing the inertia dor 1 to 20 clusters
list1= list()
for i in range(1,21):
     kmeans = KMeans( n_clusters = i)
     kmeans.fit(data)
     print('For Cluster {} the intertia is: {}'.format(i,kmeans.inertia_))
     list1.append(kmeans.inertia_)
    
#plotting line plot for no. of clusters vs inertia
frame = pd.DataFrame({'Cluster':range(1,21), 'Intertia':list1})
plt.title('Optimizing the inertia method for choosing the number of clusters',color='b')
plt.plot(frame['Cluster'],frame['Intertia'])
plt.xlabel('Number of Clusters')
plt.ylabel('Intertia')
plt.show()

#Building and fitting the kmeans model
kmeans = KMeans(n_clusters= 6)
kmeans.fit(data) 

#insert new data here for pediction
prediction= kmeans.predict(data)


```
